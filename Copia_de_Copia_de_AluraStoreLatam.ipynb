#1. Análisis de facturación



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)
# Calculamos la facturación total sumando la columna de ventas de cada tienda
facturacion_t1 = tienda['Precio'].sum()
facturacion_t2 = tienda2['Precio'].sum()
facturacion_t3 = tienda3['Precio'].sum()
facturacion_t4 = tienda4['Precio'].sum()

# Creamos una lista con los nombres de las tiendas y sus totales para compararlas
nombres_tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
totales = [facturacion_t1, facturacion_t2, facturacion_t3, facturacion_t4]

# Mostramos los resultados
for nombre, total in zip(nombres_tiendas, totales):
    print(f"La facturación total de la {nombre} es: ${total:,.2f}")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(5, 3))
plt.bar(nombres_tiendas, totales, color=['blue', 'green', 'orange', 'red'])
plt.title('Comparativa de Facturación Total por Tienda')
plt.xlabel('Tiendas')
plt.ylabel('Facturación en Pesos')
plt.show()
```

# 2. Ventas por categoría

```python
# Analizamos la frecuencia de cada categoría en las 4 tiendas
cat_tienda1 = tienda['Categoría del Producto'].value_counts()
cat_tienda2 = tienda2['Categoría del Producto'].value_counts()
cat_tienda3 = tienda3['Categoría del Producto'].value_counts()
cat_tienda4 = tienda4['Categoría del Producto'].value_counts()

print("\n--- Categorías en Tienda 1 ---")
display(cat_tienda1)
print("\n--- Categorías en Tienda 2 ---")
display(cat_tienda2)
print("\n--- Categorías en Tienda 3 ---")
display(cat_tienda3)
print("\n--- Categorías en Tienda 4 ---")
display(cat_tienda4)
```

# 3. Calificación promedio de la tienda


```python
# Calculamos la calificación promedio (rating) de cada tienda
eval_t1 = tienda['Calificación'].mean()
eval_t2 = tienda2['Calificación'].mean()
eval_t3 = tienda3['Calificación'].mean()
eval_t4 = tienda4['Calificación'].mean()

print("--- Promedio de Evaluación por Tienda (Escala 1-5) ---")
print(f"Tienda 1: {eval_t1:.2f} ⭐")
print(f"Tienda 2: {eval_t2:.2f} ⭐")
print(f"Tienda 3: {eval_t3:.2f} ⭐")
print(f"Tienda 4: {eval_t4:.2f} ⭐")
```

# 4. Productos más y menos vendidos

```python
# Función para obtener el top y bottom de cada tienda
def analizar_productos(df, nombre_tienda):
    mas_vendido = df['Producto'].value_counts().idxmax()
    cant_mas = df['Producto'].value_counts().max()

    menos_vendido = df['Producto'].value_counts().idxmin()
    cant_menos = df['Producto'].value_counts().min()

    print(f"--- {nombre_tienda} ---")
    print(f"✅ Más vendido: {mas_vendido} ({cant_mas} unidades)")
    print(f"❌ Menos vendido: {menos_vendido} ({cant_menos} unidades)\n")

# Ejecutamos el análisis para las 4 tiendas
analizar_productos(tienda, "Tienda 1")
analizar_productos(tienda2, "Tienda 2")
analizar_productos(tienda3, "Tienda 3")
analizar_productos(tienda4, "Tienda 4")
```

```python
# Calculamos el promedio de la columna de costos de envío
envio_t1 = tienda['Costo de envío'].mean()
envio_t2 = tienda2['Costo de envío'].mean()
envio_t3 = tienda3['Costo de envío'].mean()
envio_t4 = tienda4['Costo de envío'].mean()

print("--- Costo Promedio de Envío por Tienda ---")
print(f"Tienda 1: ${envio_t1:.2f}")
print(f"Tienda 2: ${envio_t2:.2f}")
print(f"Tienda 3: ${envio_t3:.2f}")
print(f"Tienda 4: ${envio_t4:.2f}")
```
